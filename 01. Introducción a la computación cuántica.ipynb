# **¿Qué es la computación cuántica?**

Modelo de programación que aprovecha propiedades cuánticas como la superposición y el entrelazamiento para realizar computaciones

  - Basado en el concepto de cúbit (_qubit_) como unidad básica de información


### Cúbits<a id="cubits"></a>

Sistema cuántico de dos estados, denotados usualmente como $|0\rangle$ y $|1\rangle$

  - El cúbit puede ponerse en superposición de ambos estados
  
El número de estados del sistema aumenta **exponencialmente** con el número de cúbits

  - 1 cúbit $\rightarrow$ 2 estados
  - 2 cúbits $\rightarrow$ 4 estados
  - 3 cúbits $\rightarrow$ 8 estados
  - $\ldots$
  - $n$ cúbits $\rightarrow$ $2^n$ estados
  
275 cúbits almacenan más estados que átomos en el universo ($2^{275} \approx 6\times 10^{82}$)

#### Emuladores (o simuladores)

Permiten emular la ejecución de un programa cuántico en un sistema clásico.
  - Necesitan gran cantidad de recursos (principalmente, memoria): uso de GPUs o sistemas HPC
  - Permiten simular sistemas de unas pocas decenas de cúbits
  - Incluyen modelos de ruido
  - Sistemas dedicados: pe. [Eviden Qaptiva-HPC](https://eviden.com/solutions/quantum-computing/qaptiva-hpc/), [Fujitsu Digital Annealer](https://www.fujitsu.com/global/services/business-services/digital-annealer/)
  - Librarías como Qiskit, Cirq, Q#, ProjectQ, etc. incorporan emuladores
  - Otros ejemplos: [Quirk](https://algassert.com/quirk), [QuEST](https://quest.qtechtheory.org/), [Qulacs](https://qulacs.org), etc.
  
Más información: https://www.quantiki.org/wiki/list-qc-simulators

# **Introducción a Qiskit**

[Qiskit](https://www.ibm.com/quantum/qiskit) (_Quantum Information Software Kit_): solución open-source, basada en Python, para crear y ejecutar programas cuánticos.

Componentes de Qiskit:

<!--center><img src="https://drive.google.com/uc?export=view&id=199hec61GLU2ZaS9r0KzHTULVPfjmNqI8" alt="Componentes de Qiskit" width="900"  /></center>

<p></p-->

 1. [Qiskit SDK](https://pypi.org/project/qiskit/)
    - Base de Qiskit: permite la definición y ejecución de programas cuánticos
    - Permite interactuar con los simuladores de Aer o sistemas reales de [IBM Quantum](https://quantum.cloud.ibm.com/) u otros (AQT, IonQ,...)

 2. [Qiskit Runtime](https://github.com/Qiskit/qiskit-ibm-runtime):
    - Entorno de ejecución optimizado para acelerar el procesamiento de programas cuánticos
    - Acelera tareas que requieren múltiples interacciones entre un procesador cuántico y uno clásico
    - Tres modos de ejecución: [Job, Session y Batch](https://quantum.cloud.ibm.com/docs/guides/execution-modes)
    - Incluye técnicas avanzadas de mitigación de errores
 2. [Qiskit Ecosystem](https://www.ibm.com/quantum/ecosystem)
    - Conjunto de herramientas y soluciones que extienden Qiskit

1. [qiskit.circuit.library](https://quantum.cloud.ibm.com/docs/api/qiskit/circuit_library). Colección de circuitos y bloques básicos de algoritmos
1. [qiskit.quantum_info](https://quantum.cloud.ibm.com/docs/api/qiskit/quantum_info): toolkit para trabajar con estados cuánticos y operadores usando colculos exactos.

1. [qiskit.primitives](https://quantum.cloud.ibm.com/docs/api/qiskit/primitives): implementaciones de referencia de las [primitivas](https://quantum.cloud.ibm.com/docs/en/guides/primitives) Sampler y Estimator
